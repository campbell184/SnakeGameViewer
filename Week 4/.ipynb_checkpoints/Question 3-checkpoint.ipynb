#### Question 3b

###### Train a NB classifier on a random 80% of the data, and use the remaining 20% to generate a confusion matrix showing its performance

```python
#Store the data in a matrix
file = open('banknote_authentication.csv')
line = file.readline()

#read all the data into data list
data = list()

line = file.readline() # removes the first line
while(line):
    data.append(line)
    line = file.readline()
    
#print(data)
```

```python
#take random 80% of the data for training
train_data = np.random.choice(data, 80, replace = False)
#print(train_data)


#remaining 20% for testing
test_data = [line for line in data if line not in train_data]
#print(test_data)
```

```python
#Split the training data into x_train_data and y_train_label
x_train_data = [line.rstrip('\n').split(';')[:-1] for line in train_data]
y_train_label = [line.rstrip('\n').split(';')[-1] for line in train_data]



#changing the arrays to floating point
x_train_data = np.array(x_train_data).astype(np.float64)
y_train_label = np.array(y_train_label).astype(np.float64)


#print('x Train data point\n')
#print(x_train_data)

#print('\ny train label\n')
#print(y_train_label)
```

```python
#Prior Probability of label y
def prior_prob(y):
    
    number_of_y = 0
    
    for label in y_train_label:
        if label == y:
            number_of_y += 1
            
    return number_of_y / len(y_train_label)

#print(prior_prob(0))
```

```python
#mean function of feature and class
def mean(feature, label):
    array = [vec[feature] for vec in x_train_data if y_train_label[np.where(x_train_data == vec)[0][0]] == label]
    
    return np.sum(array) / len(array)


#print(mean(1, 1))

#variance of feature and class
def var(feature, label):
    array = [vec[feature] for vec in x_train_data if y_train_label[np.where(x_train_data == vec)[0][0]] == label]
    result = 0
    Mean = mean(feature, label)
    
    for value in array:
        result += pow(value - Mean, 2)
        
    return result / len(array)

#print(var(1, 1))
```

```python
#conditional probability of feature given label
def con_prob(feature_value, feature, label):
    Mean = mean(feature, label)
    Var = var(feature, label)
    denominator = sqrt(2 * pi * Var)
    natural_log = exp(-0.5 * (pow(feature_value - Mean, 2) / Var) )
    probability = (1 / denominator) * natural_log
    
    return probability
```

```python
#Testing the NB Model
#input : array data_point
def Predicted_class(data_point):
    
    d = dict()
    #first - forged
    #second - genuine
    class_labels = np.arange(2)
    
    #likelihood probabilies
    #key - probability_of_x_given_label
    for label in class_labels:
        d[label] = 1
    
    
    #computing the probabilities
    for i in range(len(data_point)):
        for label in d:
            d[label] *= con_prob(data_point[i], i, label)
    
    
    denominator = 0
    for label in d:
        denominator += d[label]*prior_prob(label)
    
    #now it stores the probability_of_label_given_x
    for label in d:
        d[label] = (d[label]*prior_prob(label) / denominator)
    
    
    return [label for label in d.keys() if d[label] == max(d.values())][0]
    
    
```

```python
#splitting the test data
x_test_data = [line.rstrip('\n').split(';')[:-1] for line in test_data]
y_test_label = [line.rstrip('\n').split(';')[-1] for line in test_data]

x_test_data = np.array(x_test_data).astype(np.float64)
y_test_label = np.array(y_test_label).astype(np.float64)

#print('x Test data point\n')
#print(x_test_data)

#print('\ny Test label\n')
#print(y_test_label)

predicted_class = list()

for vec in x_test_data:
    predicted_class.append(Predicted_class(vec))

#print(predicted_class)

#computing confusion matrix
class_labels = [0, 1]
confusion_matrix = list()

count = 0

for label1 in class_labels:
    l = list()
    for label2 in class_labels:
        for i in range(len(y_test_label)):
            if predicted_class[i] == label2 and y_test_label[i] == label1:
                count += 1
        l.append(count)
        count = 0
    confusion_matrix.append(l)
    l = list()
print('Confusion Matrix\n')
print(confusion_matrix)


count_correct_match = 0
for i in range(len(predicted_class)):
    if predicted_class[i] == y_test_label[i]:
        count_correct_match += 1

print('\nAccuracy of the data: ', count_correct_match / len(y_test_label))
```
